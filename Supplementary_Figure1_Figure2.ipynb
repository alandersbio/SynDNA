### Evaluation of 16S synthetic spike-ins for shotgun metagenomics

```r
library(reshape2)
library(ggplot2)
library(ggpmisc)
library(RColorBrewer)
```

#### Run script inside directory with sequencing log files

```r
setwd("/home/zaramela/Dropbox/postdoc/metagenome_control/figures/fig1/log_files_isolates/")
```

```r
logFiles <- list.files(pattern = "*.log$")
```

```r
Treads = Fnames = Conf = NULL

for(i in 1:length(logFiles)){
    logisolate <- readLines(logFiles[i])
    totalreads <- gsub(" reads.*", "", logisolate[grep(" reads; of these", logisolate)])
    if(length(totalreads) == 0){totalreads <- 0}
    Treads <- c(Treads, totalreads)
    filenames <-  gsub("\\.fastq.*", "", logFiles[i])
    Fnames <- c(Fnames, filenames)
    conf <- gsub(".fastq", "", gsub(".log", "", gsub(".*.trimmed.fastq_", "", logFiles[i])))
    Conf <- c(Conf, conf)
}
```

```r
df <- data.frame(FileName = Fnames,
                 TotalReads = Treads,
                 Configuration = Conf)
```

```r
fwdrev <- rep("singleend", length(df$FileName))
fwdrev[grep("_1", df$FileName)] <- "foward"
fwdrev[grep("_2", df$FileName)] <- "reverse"
df$FwdRev <- fwdrev
df$FileConf <- paste0(df$FileName, "_", df$Configuration)
```

#### Loading frequency tables - Isolates

```r
setwd("/home/zaramela/Dropbox/postdoc/metagenome_control/figures/fig1/")
filenames <- list.files(pattern = "*.biom.txt$")
```

```r
dfisolates <- read.csv("isolates_table.biom.txt",header=T,sep="\t")
dfmelt <- melt(dfisolates)
dfmelt$FileName <- gsub(".fastq.trimmed.*", "", dfmelt$variable)
dfmelt$Configuration <- gsub(".fastq", "", gsub(".*.trimmed.fastq_", "", dfmelt$variable))
dfmelt$FileConf <- paste0(dfmelt$FileName, "_", dfmelt$Configuration)
dfmerge <- merge(df, dfmelt, by="FileConf")
dfmerge$Normalized <- as.numeric(dfmerge$value)/as.numeric(dfmerge$TotalReads)
dfmerge$Isolate <- gsub("_.*", "", dfmerge$FileName.x)
```

```r
notrim <- dfmerge[grep("trim50", dfmerge$Configuration.x),]
notrim <- notrim[-grep("_2", notrim$FileName.x),]
notrim16 <- notrim[-grep("synDNA", notrim$Configuration.x),]
```

```r
options(repr.plot.width=18, repr.plot.height=6)

ggplot(notrim16, aes(x = Isolate, y = sqrt(Normalized))) +
      geom_bar(stat = "identity", color="black", fill="white") +
      #geom_boxplot(outlier.shape = NA) +
      #geom_jitter(aes(color = OTUID), size = 1.5) +
      facet_wrap(~OTUID, ncol = 6) +
      coord_cartesian(ylim=c(0, 9)) +
      labs(x = "", y = "sqrt(Normalized counts)") +
      theme_bw() + guides(fill=guide_legend(ncol=1))  +
      theme(panel.border = element_rect(colour = "black", fill=NA, size=1),
            axis.text.x = element_text(angle = 90, hjust = 1, vjust = 0.5, size = 16, face = "bold", colour = "black"),
            axis.text.y = element_text(angle = 0, size = 16, face = "bold", colour = "black"),
            axis.title=element_text(size=20, face = "bold", colour = "black"),
            strip.text.x = element_text(size = 16, face = "bold", colour = "black"),
            strip.background = element_rect(fill="white"),
            legend.title=element_blank(), 
            legend.key = element_blank(),
            legend.text=element_text(size=16, face = "bold", colour = "black"))
```

#### Processing shotgun metagenomic data (ocean, soil, gut, saliva, and skin)

```r
marinelog <- readLines("logfiles/marine_log50.txt")
totalReads <- gsub(" reads.*", "", marinelog[grep(" reads; of these", marinelog)])
FileName <- gsub("_.*", "", gsub(".*log50/", "", marinelog[grep("==>", marinelog)]))
percalign <- gsub("%.*", "", marinelog[grep(" overall alignment rate", marinelog)])
alignReads <- round(as.numeric(totalReads)*(as.numeric(percalign)/100))
source <- "Marine Sponge"

dfsourcesM <- data.frame(Source = source, 
                        FileName = FileName,
                        TotalReads = totalReads,
                        ReadsAlign = alignReads,
                        Percentage = percalign,
                        CPM = (as.numeric(alignReads)/as.numeric(totalReads))*1000000)
```
